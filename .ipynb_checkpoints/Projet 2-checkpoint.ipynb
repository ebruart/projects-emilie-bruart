```python
from skimage.io import imread,imsave,imshow
from matplotlib import pyplot as plt
from skimage.color import rgb2hsv, hsv2rgb
import numpy as np
from skimage.filters.rank import mean, median
from skimage.morphology import square 

#this project has been tested with the images balloons_noisy.png, nyc.jpg and etretat.jpg, shadow.jpg

%matplotlib inline

im_init = imread("nyc.jpg")
im = imread("nyc.jpg")

#1 remove the noise with a median filter, for each color of the image, the mean filter keeps the sharp edges  

r = im[:,:,0]
g = im[:,:,1]
b = im[:,:,2]

s = square(4)

im[:,:,0]=median(r,s)
im[:,:,1]=median(g,s)
im[:,:,2]=median(b,s)


plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.imshow(im_init)
plt.title('avant')
plt.subplot(2,2,2)
plt.imshow(im)
plt.title('après')
plt.show()


#2 increase the image saturation, the colors will then seem brighter 

hsv = rgb2hsv(im)

sat = hsv[:, :, 1]

for i in range(im.shape[0]) : 
    for j in range(im.shape[1]) : 
        if sat[i,j]*1.5 > 1 : 
            sat[i,j] = 1
        else : 
            sat[i,j] *=1.5

im = hsv2rgb(hsv)

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.imshow(im_init)
plt.title('avant')
plt.subplot(2,2,2)
plt.imshow(im)
plt.title('après')
plt.show()


#2 increase the image value - choose the coeff depending on the actual value of the image 

def cumul_hist(hist):
    b=0
    cumulist = np.zeros(256)
    for i in range(len(hist)):
        b += hist[i]
        cumulist[i] = b
        
    return cumulist

def lutetirement(Tmin, Tmax):
    lut = np.arange(256)
    for i in range(256):
        if i <= Tmin :
            lut[i] = 0
        elif i >= Tmax : 
            lut[i] = 255
        else :
            m = 255/(Tmax-Tmin)
            lut[i] = int(m*(i-Tmin))
    #could be better written with the np.array syntax
    return lut

def quartile(chist, percentage) :
    gl = 0
    for i in range(chist.shape[0]) : 
        if chist[i]<(percentage*chist[255]) :
            gl = i
        else : 
            return gl

val = hsv[:,:,2]        
valint = (val*255).astype(np.uint8)


imflat = valint.flatten()
histnorm, autretruc = np.histogram(imflat, 256, density = True) #density = tru gives the cumulative hist 
plt.plot(histnorm)
plt.show()
    
c_hist = cumul_hist(histnorm)

lut = lutetirement(quartile(c_hist, 0.1), quartile(c_hist, 0.9)).astype('uint8')

new_valint = lut[valint]

imflat = new_valint.flatten()
histnorm, autretruc = np.histogram(imflat, 256, density = True) #density = tru gives the cumulative hist 
plt.plot(histnorm)
plt.show()

hsv[:, :, 2] = new_valint/255

im = hsv2rgb(hsv)

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.imshow(im_init)
plt.title('avant')
plt.subplot(2,2,2)
plt.imshow(im)
plt.title('après')
plt.show()
```
